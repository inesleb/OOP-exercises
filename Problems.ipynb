# Operator Overloading

## Problem 1

```python
class Complex:
    def __init__(self, real, imaginary):
        self.real = real
        self.imaginary = imaginary

    def __sub__(self, other):
        return Complex(
            self.real - other.real,
            self.imaginary - other.imaginary)
        
    def __repr__(self):
        return f"{self.real} + {self.imaginary}i"
```

```python
a = Complex(3,3)
b = Complex(5,2)
```

```python
a.__sub__(b)
```

```python
Complex.__sub__(a,b)
```

```python
a - b
```

```python
print(a.__sub__(b))
```

```python
print(a - b)
```

```python
class Person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender
        
    def __gt__(self, other): # > operator
        if self.age > other.age:
            return True
        else:
            return False
```

## Problem 5

```python
class Circle:
    def __init__(self, radius):
        self.radius = radius
        
    def __gt__(self, other): # > operator
        if self.radius > other.radius:
            return True
        else: 
            return False

    def __lt__(self, other): # < operator
        if self.radius < other.radius:
            return True
        else: 
            return False

    def __eq__(self, other): # == operator
        if self.radius == other.radius:
            return True
        else:
            return False

    def __ne__(self, other): # != > operator
        if self.radius != other.radius:
            return True
        else:
            return False

    def __ge__(self, other): # >= operator
        if self.radius >= other.radius:
            return True
        else:
            return False

    def __le__(self, other): # <= operator
        if self.radius <= other.radius:
            return True
        else:
            return False
```

```python
c1 = Circle(10)
c2 = Circle(15)
c3 = Circle(15)
```

```python
c1 > c2
```

```python
c2 >= c2
```

## Problem 6

```python
import math

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
            
    def __gt__(self, other): # > operator
        if math.sqrt(self.x ** 2 + self.y ** 2) > math.sqrt(other.x ** 2 + other.y ** 2):
            return True
        else: 
            return False

    def __lt__(self, other): # < operator
        if math.sqrt(self.x ** 2 + self.y ** 2) < math.sqrt(other.x ** 2 + other.y ** 2):
            return True
        else: 
            return False

    def __eq__(self, other): # == operator
        if math.sqrt(self.x ** 2 + self.y ** 2) == math.sqrt(other.x ** 2 + other.y ** 2):
            return True
        else:
            return False

    def __ne__(self, other): # != > operator
        if math.sqrt(self.x ** 2 + self.y ** 2) != math.sqrt(other.x ** 2 + other.y ** 2):
            return True
        else:
            return False

    def __ge__(self, other): # >= operator
        if math.sqrt(self.x ** 2 + self.y ** 2) >= math.sqrt(other.x ** 2 + other.y ** 2):
            return True
        else:
            return False

    def __le__(self, other): # <= operator
        if math.sqrt(self.x ** 2 + self.y ** 2) <= math.sqrt(other.x ** 2 + other.y ** 2):
            return True
        else:
            return False
```

```python
p1 = Point(1, 1)
p2 = Point(2, 2)
p3 = Point(2, 2)
```

```python
p1 > p2
```

```python
p2 <= p3
```

## Problem 7

```python
class Matrix:
    def __init__(self, data):
        self.data = data
        self.rows = len(data)
        self.cols = len(data[0])

    def __mul__(self, other): # * operator
        # Check 1
        if not isinstance(other, Matrix):
            raise TypeError("Can only multiply by another Matrix")

        # Check 2
        if self.cols != other.rows:
            raise ValueError("Incompatible dimensions for matrix multiplication")

        # Result
        result = [
            [0 for _ in range(other.cols)]
            for _ in range(self.rows)
        ]

        # Computations
        for i in range(self.rows):
            for j in range(other.cols):
                for k in range(self.cols):
                    result[i][j] += self.data[i][k] * other.data[k][j]

        return Matrix(result)

    def __repr__(self):
        return f"Matrix({self.data})"
```

```python
m1 = Matrix([[1, 2], [3, 4]])
m2 = Matrix([[2, 1], [4, 3]])
```

```python
m1 * m2
```

# Inheritance

## Problem 1

```python
class Vehicule:
    def __drive__(self):
        return("Driving a vehicule.")

class Car(Vehicule):
    def __drive__(self):
        return("Driving a car.")

class Bike(Vehicule):
    def __drive__(self):
        return("Driving a bike.")
```

```python
car = Car()
bike = Bike()
print(car.__drive__())
print(bike.__drive__())
```

## Problem 2

```python
class Person:
    def __init__(self, name, age, address):
        self.name = name
        self.age = age
        self.address = address

    def introduce(self):
        return f"My name is {self.name}, I am {self.age} and live at {self.address}."

        
class Student(Person):
    def __init__(self, name, age, address, field_of_study):
        Person.__init__(self, name, age, address)
        self.field_of_study = field_of_study

    def introduce(self):
        introduction = Person.introduce(self)
        return f"{introduction} I study {self.field_of_study}."


class Employee(Person):
    def __init__(self, name, age, address, company):
        Person.__init__(self, name, age, address)
        self.company = company

    def introduce(self):
        introduction = Person.introduce(self)
        return f"{introduction} I work at {self.company}."
```

```python
p1 = Student("Lina", 20, "1 rue Jolie", "Engineering")
p2 = Employee("Mandy", 34, "2 rue Dodo", "Bic&Co")
print(p1.introduce())
print(p2.introduce())
```

## Problem 3

```python
class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance
        
    def deposit(self, amount):
        self.balance += amount
        
    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            return True
        else:
            return False
    
    def transfer(self, destination_account, amount):
        if self.balance >= amount:
            self.balance -= amount
            destination_account.deposit(amount)
            return True
        else:
            return False

class SavingsAccount(BankAccount):
    def __init__(self, account_number, balance, interest_rate):
        BankAccount.__init__(self, account_number, balance)
        self.interest_rate = interest_rate
        
    def add_interest(self):
        interest = self.balance * self.interest_rate
        self.deposit(interest)
        
    def check_balance(self):
        return self.balance

savings_account1 = SavingsAccount(account_number="12345", balance=1000, interest_rate=0.05)
savings_account2 = SavingsAccount(account_number="67890", balance=500, interest_rate=0.03)

# transfer money from one account to another
if savings_account1.transfer(destination_account=savings_account2, amount=100):
    print("Transfer successful")
else:
    print("Transfer failed")

# check the balance of both accounts
print("Savings Account 1 Balance:", savings_account1.check_balance())
    
```
